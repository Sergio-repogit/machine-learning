Un árbol de decisión en el contexto de ingeniería informática se construye mediante un proceso de aprendizaje automático que segmenta el espacio de datos en subconjuntos más manejables para abordar problemas de software o hardware. En cada nodo del árbol, se selecciona la característica que divide el conjunto de datos según la variable objetivo. Por ejemplo, en el desarrollo de software, podría usarse para decidir entre diferentes algoritmos de codificación basándose en criterios como complejidad de implementación, eficiencia de tiempo de ejecución y uso de recursos.

Luego, los árboles de decisión son una técnica explicativa y descomposicional que utiliza un proceso de división secuencial, iterativo y descendente que, partiendo de una variable dependiente, forma grupos homogéneos definidos específicamente mediante combinaciones de variables independientes en las que se incluyen la totalidad de los casos recogidos en la muestra.

Esta técnica se basa en una estructura de árbol en la que cada nodo representa una "pregunta" o prueba sobre una característica (por ejemplo, ¿el tiempo de ejecución del algoritmo es menor que el tiempo promedio requerido?), cada rama representa el resultado de la prueba, y cada hoja representa una etiqueta de clase (el resultado final de la clasificación). Los árboles de decisión dividen el espacio de datos basándose en el concepto de entropía de la teoría de la información. La ganancia de información se utiliza para medir cuánto ayuda una característica a reducir la incertidumbre o impureza en los datos.

Al evaluar una posible división de un nodo, se calcula la entropía para cada subconjunto resultante y se ponderan estas entropías por el tamaño de los subconjuntos para obtener la entropía promedio después de la división. La ganancia de información es entonces la diferencia entre la entropía original del nodo y esta entropía promedio ponderada. El atributo o variable que resulte en la máxima ganancia de información será elegido para dividir el nodo. Este proceso se repite recursivamente hasta que se alcanzan los criterios de terminación del árbol, como un número mínimo de muestras en cada nodo o una ganancia de información mínima. 

Las principales características de los árboles de decisión incluyen su interpretabilidad, la no necesidad de asumir una distribución específica de los datos, la capacidad para manejar datos heterogéneos, y su rápida evaluación, lo que es crucial en entornos donde las decisiones deben tomarse rápidamente. Sin embargo, son susceptibles al sobreajuste y sensibles a cambios en los datos de entrada, por lo que técnicas como la poda del árbol y la validación cruzada son esenciales para mantener su efectividad.

Las máquinas de vector soporte (SVM) son un modelo de aprendizaje supervisado que se utiliza para clasificación. El objetivo principal de un SVM es encontrar un hiperplano en un espacio N-dimensional (N — el número de características) que clasifique de manera óptima los puntos de datos en dos categorías. Esto se logra maximizando el margen entre las clases de datos.

En este sentido, las máquinas de vector soporte son útiles para problemas de clasificación binaria en ingeniería informática, como el filtrado de spam, por ejemplo, debido a su capacidad para crear modelos que no solo son precisos, sino también capaces de generalizar bien en situaciones de datos reales y complejos. Si suponemos que estamos diseñando un sistema para clasificar emails automáticamente en "spam" o "no spam". En este escenario, cada email es un punto en nuestro espacio N-dimensional donde las dimensiones podrían incluir características como la frecuencia de ciertas palabras, la longitud del email, el uso de mayúsculas, entre otras.

Los modelos de SVM buscarán el hiperplano que deja el máximo margen posible entre los correos clasificados como spam y los que no lo son. La distancia entre el hiperplano y los puntos más cercanos de cada clase (conocidos como vectores de soporte) es maximizada.

En casos donde los datos no sean linealmente separables (p.ej., características de emails que no permitan una división clara usando un hiperplano), los modelos SVM utilizarán una función kernel para transformar los datos a un espacio de mayor dimensión donde la separación lineal sea posible. Kernels comunes incluyen el polinomial y el de base radial (RBF) que nosotros no vamos a ver en esta asignatura.

Las principales características de los modelos SVM que podemos destacar serían las siguientes: modelos adecuados para problemas de clasificación binaria y requieren que los datos estén normalizados para que la escala de las características no sesgue el hiperplano. Aunque el entrenamiento puede ser intensivo computacionalmente, una vez entrenado, el modelo es eficiente y efectivo al generalizar sobre datos no vistos, lo que es crucial para aplicaciones como la detección de spam en tiempo real. La maximización del margen y técnicas de regularización ayudan a hacer de SVM un modelo robusto, especialmente útil en entornos con alto riesgo de sobreajuste.

El método de los k vecinos más cercanos (KNN) es un algoritmo simple y eficaz usado en problemas de clasificación en machine learning. El modelo KNN asigna la etiqueta de clase a una muestra basándose en la prevalencia de etiquetas entre sus k vecinos más cercanos dentro del conjunto de entrenamiento. Este método se basa en el principio de que las muestras similares tienden a estar cercanas en el espacio de características.

El funcionamiento del KNN comienza con la elección de un número k, que es el número de vecinos más cercanos a considerar para la clasificación de una nueva muestra. El algoritmo calcula la distancia (usualmente la distancia euclidiana) entre la muestra nueva y todas las muestras en el conjunto de entrenamiento, seleccionando las k más cercanas. La clasificación se determina por mayoría de votos entre los k vecinos: la clase que más aparece entre los k vecinos es asignada a la muestra nueva.

Por ejemplo, si suponemos que una empresa de tecnología desea implementar un sistema automático para clasificar los correos electrónicos recibidos en categorías como *urgente*, *importante*, *normal* y *spam*. Cada correo se representa en un espacio de características que incluye elementos como la longitud del texto, frecuencia de palabras clave, tiempo de envío, y otros metadatos.

En este caso, los correos electrónicos se convertirán en vectores de características. Esto puede incluir la codificación de texto a través de técnicas como TF-IDF, que mide la importancia de una palabra en relación con un conjunto de documentos que vimos en el bloque anterior de la asignatura. Se elige un k adecuado, usualmente a través de validación cruzada, para equilibrar entre hacer el modelo demasiado general o demasiado sensible a los ruidos en los datos de entrenamiento. Para un nuevo correo, se calculan las distancias a todos los correos en el conjunto de entrenamiento. Las k distancias más cortas determinan los vecinos más cercanos. El nuevo correo se clasifica según la categoría que más predomina entre sus k vecinos más cercanos.

Las principales características de los modelos KNN que podemos destacar serían las siguientes: es intuitivamente simple y fácil de implementar. No asume nada sobre la forma de los datos, lo que lo hace flexible y adaptable. Al ser un método no paramétrico, KNN es particularmente útil cuando no hay conocimiento previo sobre la distribución de los datos. Como algoritmo *perezoso*, KNN no construye un modelo explícito del conjunto de entrenamiento. En cambio, utiliza el conjunto de entrenamiento completo para la clasificación durante la fase de test, lo que puede resultar en un alto costo computacional y de almacenamiento. La elección de k afecta significativamente al rendimiento del algoritmo. Un k muy pequeño puede hacer el modelo susceptible al ruido, mientras que un k muy grande lo hace vulnerable a las clasificaciones erróneas debido a la inclusión de muestras irrelevantes.

Cada uno de los tres modelos vistos tiene un hiperparámetro que sirve para poder optimizar durante el proceso de entrenamiento el modelo que se esté entrenando. A continuación, detallamos las principales catacterísticas de cada uno de ellos.

* **Árbol de decisión.** El valor del hiperparámetro `cp` (parámetro de complejidad) controla el tamaño del árbol, donde valores más pequeños permiten árboles más complejos (más grandes) y valores más grandes conducen a árboles más pequeños (simplificados) de entre todos los posibles árboles (nodos-ramas-hojas) que pudieran existir. Recordar que cada algoritmo de clasificación tiene sus propios hiperparámetros con diferente significado.

------------

* **Máquinas de Soporte Vectorial.** El valor `C` es un parámetro de regularización que controla la penalización por error de clasificación y el margen suave en las SVM lineales. En SVM lineales, el objetivo es encontrar el hiperplano de separación que maximiza el margen entre las clases en el espacio de características. El parámetro `C` controla el equilibrio entre dos objetivos:

   - <u>Minimizar el margen</u>: un `C` más grande penalizará más los errores de clasificación y buscará un hiperplano de separación que separe correctamente las clases, incluso si eso significa tener un margen más estrecho.

   - <u>Maximizar el margen</u>: un `C` más pequeño permitirá más errores de clasificación en el conjunto de entrenamiento, pero buscará un hiperplano de separación que tenga un margen más amplio, lo que puede resultar en una mejor generalización a datos nuevos.

------------

* **K-vecinos cercanos.** El valor `k` especifica cuántos de los vecinos más cercanos en el espacio de características se deben considerar para determinar la salida de una nueva instancia. Por ejemplo, si $k = 5$, los cinco vecinos más cercanos a la nueva muestra se usarán para predecir la clase o valor de esa muestra. El parámetro `k` podría oscilar entre:

   - <u>`k` pequeño</u>: puede hacer al modelo extremadamente sensible al ruido presente en los datos. Con k muy pequeño, incluso anomalías o outliers pueden tener un alto impacto en la predicción, lo que potencialmente conduce a un sobreajuste.

   - <u>`k` grande</u>: hace que el modelo sea más robusto frente al ruido y menos sensible a outliers, pero puede llevar a un modelo demasiado generalizado (subajuste), donde las peculiaridades en los datos no se capturan bien.


En resumen, para interpretar el resultado, es importante considerar cómo se selecciona el valor de `cp`, `C` o `k`. En la función **`train()`** de la librería $\tt caret$, generalmente se realiza una búsqueda en una cuadrícula de valores para dichos coeficientes, y se selecciona el valor que maximiza la precisión o alguna otra métrica de rendimiento aplicando la técnica de grid-search. Los valores que seleccionamos de la cuadrícula dependerá del tipo de hiperparámetro (modelo) que estemos empleando. Por ejemplo, para entrenar un modelo KNN, usualmente definimos un rango de valores de k para explorar, normalmente entre $0$ y $15$-$20$. Después, a través de la librería $\tt caret$ evaluaremos el rendimiento del modelo para cada valor de k dentro del rango especificado usando validación cruzada, por ejemplo, y luego, seleccionará el valor de k que mejor rendimiento tenga según la métrica escogida (por ejemplo, la precisión). Esto se configura en el argumento `tuneGrid`.

#### <mark>Entrenamiento del modelo SVM aplicando la técnica de grid search</mark>

En este caso, nos piden que utilicemos la técnica de grid search para optimizar los parámetros del modelo. Para ello, consideraremos el argumento `expand.grid(C = c(0.1, 1, 10))` que define los valores que se probarán para el parámetro `C` durante la búsqueda en la cuadrícula. Recordar que el valor `C` es un parámetro de regularización que controla la penalización por error de clasificación y el margen suave en las SVM lineales. En SVM lineales, el objetivo es encontrar el hiperplano de separación que maximiza el margen entre las clases en el espacio de características. El parámetro `C` controla el equilibrio entre dos objetivos (<u>repaso</u>):

   - **Minimizar el margen**: un `C` más grande penalizará más los errores de clasificación y buscará un hiperplano de separación que separe correctamente las clases, incluso si eso significa tener un margen más estrecho.


   - **Maximizar el margenn**: un `C` más pequeño permitirá más errores de clasificación en el conjunto de entrenamiento, pero buscará un hiperplano de separación que tenga un margen más amplio, lo que puede resultar en una mejor generalización a datos nuevos.


Como comentamos antes, en la función **`train()`** de la librería $\tt caret$, generalmente se realiza una búsqueda en una cuadrícula de valores para `C`, y se selecciona el valor que maximiza la precisión o alguna otra métrica de rendimiento aplicando la técnica de grid search. Esto se configura en el argumento `tuneGrid`. Veámoslo aplicado a este conjunto de datos.

```python
# Cuadrícula de parámetros para la búsqueda de hiperparámetros
grid = expand.grid(C = c(0.1, 1, 10)) 
```

#### <mark>Entrenamiento del modelo k-NN aplicando la técnica de grid search</mark>

En este caso, nos piden que utilicemos la técnica de grid search para optimizar los parámetros del modelo. Para ello, consideraremos el argumento `expand.grid(k = 1:20)` que define los valores que se probarán para el parámetro $k$ durante la búsqueda en la cuadrícula. Recordar que el valor $k$ especifica cuántos de los vecinos más cercanos en el espacio de características se deben considerar para determinar la salida de una nueva instancia. Por ejemplo, si $k = 5$, los cinco vecinos más cercanos a la nueva muestra se usarán para predecir la clase o valor de esa muestra. El parámetro *k* podría oscilar entre (<u>repaso</u>):

   - **k pequeño**: puede hacer al modelo extremadamente sensible al ruido presente en los datos. Con k muy pequeño, incluso anomalías o outliers pueden tener un alto impacto en la predicción, lo que potencialmente conduce a un sobreajuste.


   - **k grande**: hace que el modelo sea más robusto frente al ruido y menos sensible a outliers, pero puede llevar a un modelo demasiado generalizado (subajuste), donde las peculiaridades en los datos no se capturan bien.

La librería $\tt caret$ evaluará el rendimiento del modelo para cada valor de `k` dentro del rango especificado, a menudo usando validación cruzada, y luego seleccionará el valor de `k` que mejor rendimiento tenga según la métrica escogida (por ejemplo, la precisión). Esto se configura en el argumento `tuneGrid`. Veámoslo aplicado a este conjunto de datos.

```python
# Cuadrícula de parámetros para la búsqueda de hiperparámetros (valores de k de 1 a 20)
grid = expand.grid(k = 1:20) 
```
