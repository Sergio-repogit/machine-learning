<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

La idea tras la reduccion de la dimension es entender que podemos proyectar nuestros datos (que se tienen en una base de datos de dimension N) en un hiperplano (o un espacio vectorial de dimension menor) y proyectar los elementos de nuestra base de datos sobre ese hiperplano (o espacio de dimension menor).

Cuando hacemos esta representacion estamos obteniendo los elemtos en una dimension menor pero estamos perdiendo la informacion propia de los datos. La gracia es que si encontramos un buen hiperplano (las direcciones principales) podemos encontrar una pryeccion que puede mantener la mayor cantidad de informacion de la base de datos original.

Cuando estamos hablando de la informacion nos referimos a la relacion que tienen los puntos entre ellos. Las vecindades. Algun patron en los puntos que se mantenga cuando hagamos esa proyeccion.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Tenemos 3 criterios principales para determinar lo buena que es nuestra reduccion de la dimensionalidad.

+ La **`varianza explicada`**. Esta solo se puede aplicar cuando estamos en un metodo que se base en combinaciones lineales de las variables originales. Esta metrica se basa en la varianza que es capaz de explicar (la base reducida) de la base original. Este es un metodo que puede llegar a entender las bases de datos en funcion de relaciones lineales.

+ **`Error de construccion`** Este metodo se basa en los models que tengamos un metodo para recontruir la base original en funcion de la base de datos reducida. Esta metrica se calcularar haciendo una diferencia entre la base de datos original y la reconstruida. Esta no es una metrica que se puede utilizar siempre y dependera de si tenemos un metodo de recontruccion de la base de datos original. Ademas no tenemos una buena interpretabilidad de que es el fallo que estamos teniendo. Ademas es un poco raro la interpretabilidad de este error.

+ **`Preservacion de distancias o vencidarios`**. La idea es analizar las vecindades, los grupos y la distancia entre los puntos en la base original y compararlos con las mismas vecindades, grupos y distnacia entre la base de datos reducida (tras la reduccion de la dimensionalidad). Es una buena metrica a la hora de la interpretabilidad (si se pierde un patron se entiende que estamos perdiendo esa informacion) pero a la hora de la implementacion puede llegar a ser muy complicado calcularlo porque tenemos que eneocntrar esas relaciones en la original.

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

Los algoritmos de reducción de dimensionalidad pueden organizarse en distintas familias según el tipo de transformación que aplican al espacio original y la estructura que intentan preservar(varianza, distancias, vecindarios, etc.). La distinción más habitual separa entre métodos lineales y métodos no lineales, dependiendo de si la relación entre las variables originales y las nuevas dimensiones se modela mediante combinaciones lineales o mediante transformaciones más flexibles(no lineales). Vamos a analizar ambos tipos de métodos por separado.

Los modelos lineales buscan una proyección lineal de los datos a un espacio de menor dimensión. Cada nueva dimensión es una combinación lineal de las variables originales. Tienen como ventajas su simplicidad, interpretabilidad y el menor coste computacional, por lo que suelen ser la primera opción elegida o cuando se desea comprender cómo se relacionan las nuevas dimensiones con las variables iniciales. Existen tres variantes principales:

- Métodos de proyección: su objetivo es encontrar combinaciones lineales que expliquen la mayor varianza posible o que reproduzcan bien la estructura global de los datos. El ejemplo más representativo es el Análisis de Componentes Principales (PCA). PCA combina variables para explicar la varianza.


- Métodos de factores latentes: asumen que las variables observadas son el resultado de un número reducido de factores no observados más un término de error. Buscan interpretar estas dimensiones latentes. El modelo más utilizado en este caso es el Análisis Factorial (AF). AF usa factores latentes para explicar las variables.


- Métodos de variable respuesta: incorporan una variable de respuesta y construyen las nuevas dimensiones de forma que separan bien las clases o explican la respuesta. Son útiles cuando la reducción de dimensionalidad es un paso previo a un modelo predictivo. El modelo más utilizado es el Análisis Discriminante Lineal (LDA). LDA combina variables para separar las clases.


Los métodos no lineales intentan capturar estructuras más complejas en los datos, como curvas o variedades de baja dimensión inmersas en espacios de alta dimensión. En lugar de limitarse a combinaciones lineales, utilizan relaciones no lineales y grafos de vecindad. Son especialmente útiles cuando la estructura de los datos no puede describirse bien con un plano. Dentro de esta familia se distinguen tres subtipos:


- Métodos basados en distancias globales (escalamiento): parten de una matriz de distancias o disimilitudes entre observaciones y buscan una representación en baja dimensión que las respete lo mejor posible. El método más utilizado es el Multidimensional Scaling (MDS). Son adecuados cuando lo principal es preservar las distancias globales entre todos los pares de puntos.


- Métodos basados en vecindarios locales (manifold learning): asumen que los datos se encuentran cerca de una variedad de baja dimensión y construyen la representación reducida preservando las relaciones locales (quién es vecino de quién). El método más utilizado es el Isomap o Locally Linear Embedding (LLE). Son útiles cuando la estructura global es curvada o intrincada, pero localmente los datos se comportan casi de forma lineal.


- Métodos basados en grafos de similitud o distribuciones de probabilidad: construyen un grafo de proximidad o distribuciones de probabilidad sobre pares de puntos y buscan una representación en baja dimensión que preserve las probabilidades de vecindad. Son muy potentes para visualización de alta dimensión. Los modelos más utilizados son los t-distributed Stochastic Neighbor Embedding (t-SNE) y Uniform Manifold Approximation and Projection (UMAP). Destacan por resaltar bien grupos o estructuras locales, aunque la relación global entre grupos puede ser menos interpretable.



<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

Seleccionar el número óptimo de dimensiones reducidas r no es una tarea trivial. De manera análoga al número de clústeres q, aquí buscamos un compromiso entre representar bien la información y mantener un modelo simple, es decir, pocas dimensiones.

Existen tres métodos muy empleados en la práctica para determinar el mejor número de dimensiones reducidas r: el método de Kaiser en el gráfico de sedimentación, el método del porcentaje de varianza explicada y método del error de reconstrucción o rendimiento. Veamos a continuación cada uno de estos métodos.

+ Método de Kaiser: Este método es el más conocido para decidir cuántas dimensiones mantener en un problema de reducción de dimensionalidad basado en métodos lineales. Cada componente principal tiene asociado un autovalor $\lambda_k$, que mide la varianza explicada por dicha componente. Cuanto mayor es $\lambda_k$, mayor es la cantidad de información (variabilidad) que captura esa dimensión. Para encontrar el número óptimo de dimensiones $r$ se llevan a cabo los siguientes pasos:

  1. Calcular todos los autovalores $\lambda_1 \geq \lambda_2 \geq \cdots \geq \lambda_p$.


  2. Graficar en el eje horizontal el número de componentes $k = 1,...,p$ y en el eje vertical el valor de los autovalores $\lambda_k$.

  El criterio de Kaiser propone tomar como número de componentes $r$ el número de autovalores mayores que $1$, interpretando que cada componente retenida debe aportar al menos tanta información como una variable original.

  La idea intuitiva es la siguiente. Antes de extraer componentes, cada variable original está estandarizada y, por tanto, tiene varianza 1. Un autovalor $\lambda_k$ mide cuánta varianza total explica la componente $k$.

  + Si $\lambda_k > 1$, esa componente está explicando más varianza que una variable original promedio. La mantenemos porque resume tanta información o más que una variable original.


  + Si $\lambda_k < 1$ la componente explica menos varianza que una sola variable estandarizada, de modo que se considera poco eficiente mantenerla como dimensión adicional.


+  Método del porcentaje de variable explicada: El segundo criterio se basa en el porcentaje de varianza explicada acumulada por las componentes principales.


    1. Se ordenan las componentes por importancia (autovalor) y se calcula, para cada $k$, la varianza explicada acumulada por las primeras $k$ componentes.


    2. Se fija de antemano un umbral mínimo de varianza explicada.

    El número óptimo de componentes $r$ será el menor valor de $k$ tal que la varianza explicada acumulada por las primeras $k$ componentes sea mayor o igual que el umbral fijado. Es decir, escogemos el mínimo número de componentes que permita conservar, al menos, ese porcentaje de información de los datos originales.


+ Método del error de reconstrucción o rendimiento: Se basa en probar varios valores candidatos de $r$ (número de dimensiones) y ver cómo se comporta el sistema con cada uno de ellos.

  1. Se fijan varios valores candidatos de $r$.


  2. Para cada $r$:
      - Se ajusta el modelo de reducción de dimensionalidad elegido.
      - Se calcula el error de reconstrucción entre los datos originales y los reconstruidos, o el rendimiento del modelo final que usa solo esas $r$ dimensiones (MSE, MAE, RMSE, etc.).


  3. Se elige como número óptimo de dimensiones aquel $r$ que consigue un error de reconstrucción suficientemente bajo sin necesidad de mantener todas las dimensiones.

  El criterio consiste básicamente en hacer un grid search para r.

En el objeto PCA hay varios componentes que podriamos comentar pero no nos aportan informacion relevante que podamos usar.
Vamos a hacer un comentario breve de lo que significan pero no interpretaremos los valores de nuestro objeto PCA.

+ **`var`**: Nos proporciona **cuatro matrices** con los siguientes **resultados del ACP para las variables** consideradas:

    - `coord`: coordenadas de cada variable en los ejes (cargas factoriales).
    - `cor`: correlaciones variable-componente (en ACP estándar coinciden con `coord`).
    - `cos2`: calidad de representación de cada variable en cada eje (proporción de varianza de la variable explicada por ese eje).
    - `contrib`: contribución ($\%$) de cada variable a la construcción de cada componente.

+ **`ind`**: Nos proporciona **cuatro matrices** con los siguientes **resultados del ACP para los individuos** considerados:

    - `coord`: coordenadas de cada alumno en las nuevas dimensiones.
    - `cos2`: qué parte de la inercia de cada alumno queda explicada por cada eje (calidad de representación).
    - `contrib`: aportación de cada alumno a la varianza de cada componente.
    - `dist`: distancia de cada alumno al centro de gravedad en el espacio original (inercia individual).

+ **`svd`**: Nos proporciona **tres matrices** con la **descomposición en valores singulares** usada internamente por $\tt FactoMineR$ para calcular el ACP. Normalmente esta información no la usamos para interpretación “aplicada”, pero **es útil si queremos entrar al detalle algebraico**:

    - `vs`: valores singulares.
    - `U`: matriz de coordenadas normalizadas de los individuos.
    - `V`: matriz de vectores propios asociados a las variables.

+ **`call`**: Nos proporciona una **lista con nueve elementos** con la información de la llamada al modelo y el preprocesado. Estos datos sirven para **reproducir el análisis y entender exactamente cómo se ha construido** el ACP:

    - Pesos de filas (`row.w`) y columnas (`col.w`).
    - Si se estandarizaron las variables (`scale.unit`).
    - Número de componentes calculados (`ncp`).
    - Medias (`centre`) y desviaciones típicas (`ecart.type`) usadas para centrar y escalar.
    - La matriz de datos estandarizada (`X`) y la llamada original (`call`).

Todas estas variables no es que sean inutiles de por si pero se utilizan para hacer un analisis mucho mas tecnico de como nuestro modelo ha realizado el PCA.

Estaria interesante analizar Que componenetes son las mas importantes y la contribucion de cada columna (variable) o fila (observacion) que tiene en nuestro metodo pero es un analisis de mas alto nivel que no se nos pide.

Para determinar el numero optimo de componentes principales utilizaremos 3 metodos distintos:

+ `Metodo de Kaiser`: en el metodo de Kaiser consideramos todos los autovalores que son mayores que 1. Sabemos que minimo tiene que haber mas de uno debido a que todos los valores estan estandarizados. Este metodo se basa en que los autovalores mayores guardan una mayor varianza (pierden menos información)

+ `Porcentaje de varianza`: Este metodo se basa en el mimso fundamento que en el apartado anterior pero tiene otra metrica distinta. Utilizara la varianza acumulada. Tomara tantos autovalores hasya que la siguiente expresion de un valor mayor que 0.8 o 0.9 $$\frac{\sum_{i=1}^{m}\lambda_i}{\sum_{j=1}^{n}\lambda_j}$$ Con m el numero de autovalores que estariamos tomando (el numero de dimension a la que estamos reduciendo) y n el numero de variables original.

```r
# Aplicamos el criterio del método de Kaiser
fviz_eig(PCA_out, choice = "eigenvalue", geom = "bar") +
  theme_grey() +
  geom_hline(yintercept = 1, colour = "red", linetype = "dashed", linewidth = 1)
```

Hemos expicado previamente cual el criterio de Kaiser y en que se basa. Ahora recordaremos que se basa en solo tomar los autovalres con valores mayores que 1.

+ Vemos que el **único** autovalor que tomamos es el primero debido a que es el único que supera el umbral de 1 (con un valor mayor a 6)

Segun el metodo de Kaiser estamos eliminando 7 variables para tomar una representacion solo con 1 dimension ya que concentran la mayor parte de informacion de la base de datos original.


Método del porcentaje de varianza

```r
# Tabla de autovalores y porcentajes de varianza (igual que PCA_out$eig)
eig_tab = get_eigenvalue(PCA_out)
eig_tab
```

```r
# Extraemos la varianza acumulada
var_eig = data.frame(dim = 1:nrow(eig_tab), cum = eig_tab[, "cumulative.variance.percent"])
var_eig
```

```r
# Número óptimo de componentes según el criterio del porcentaje de varianza explicada (fijamos en 75%)
umbral = 85
r_opt = which(var_eig$cum >= umbral)[1]; r_opt
```

```r
# Aplicamos el criterio del método del porcentaje de varianza explicada
fviz_eig(PCA_out, choice = "variance", geom = "line") %+% var_eig +
  aes(x = dim, y = cum) +
  geom_point() +
  geom_line() +
  geom_hline(yintercept = umbral, linetype = "dashed") +
  geom_vline(xintercept = r_opt,  linetype = "dashed") +
  scale_x_continuous(breaks = var_eig$dim) +
  coord_cartesian(ylim = c(0, 100)) +
  labs(title = "Selección por porcentaje de varianza explicada",
           x = "Número de componentes (k)",
           y = "Varianza explicada acumulada (%)",
    subtitle = paste("Umbral = ", umbral, "%  →  r óptimo =", r_opt)) +
  theme_grey()
```

En este grafico se muestra la varianza acumulada por los componentes principales. Estamos mostrando tambien un umbral de `85%` (la linea discontinua horizontal)

+ Vemos que el primer componente muestra un valor un poco mayor a 75%. Aunque sea una buena cantidad para solo un componente no es un valor lo sufientemente mayor para el umbral que hemos impuesto. Tendremos que tomar mas componentes.

+ Para los dos primeros componenetes se obtiene una varianza un poco mayor a 80% pero no llega a ese umbral de 85% por lo que tomamos otro valor.

+ Cuando tomamos el 3 componente observamos como ya estamos superando ese umbral que hemos impuesto del 85%.


Nuestro metodo nos incida que tenemos que tomar 3 componentes para superar ese 85% de explicacitivilidad de la varianza. Por lo que segun este metodo (y el umbral que hemos impuesto nosotros) tomaremos `k=3` lo que significa que nos quedaremos con una nueva base de datos con 3 variables en vez de 8.

El objeto que hemos creado PCA contiene una lista con 5 elementos con la siguiente información:

+ eig: Es una matriz con la información de los autovalores de cada componente:

    - eigenvalue: valor propio (varianza explicada por cada dimensión).
    - percentage of variance: % de varianza explicada por cada componente.
    - cumulative percentage of variance: % acumulado hasta ese componente. Es la tabla que usamos para decidir cuántas dimensiones elegimos.

+ var: Nos proporciona cuatro matrices con los siguientes resultados del ACP para las variables consideradas:

    - coord: coordenadas de cada variable en los ejes (cargas factoriales).
    - cor: correlaciones variable-componente (en ACP estándar coinciden con coord).
    - cos2: calidad de representación de cada variable en cada eje (proporción de varianza de la variable explicada por ese eje).
    - contrib: contribución porcentual de cada variable a la construcción de cada componente.

+ ind: Nos proporciona cuatro matrices con los siguientes resultados del ACP para los individuos considerados:

    - coord: coordenadas de cada alumno en las nuevas dimensiones.
    - cos2: qué parte de la inercia de cada alumno queda explicada por cada eje (calidad de representación).
    - contrib: aportación de cada alumno a la varianza de cada componente.
    - dist: distancia de cada alumno al centro de gravedad en el espacio original (inercia individual).

+ svd: Nos proporciona tres matrices con la descomposición en valores singulares usada internamente por FactoMineR para calcular el ACP. Normalmente esta información no la usamos para interpretación “aplicada”, pero es útil si queremos entrar al detalle algebraico:

    - vs: valores singulares.
    - U: matriz de coordenadas normalizadas de los individuos.
    - V: matriz de vectores propios asociados a las variables.

+ call: Nos proporciona una lista con nueve elementos con la información de la llamada al modelo y el preprocesado. Estos datos sirven para reproducir el análisis y entender exactamente cómo se ha construido el ACP:

    - Pesos de filas (row.w) y columnas (col.w).
    - Si se estandarizaron las variables (scale.unit).
    - Número de componentes calculados (ncp).
    - Medias (centre) y desviaciones típicas (ecart.type) usadas para centrar y escalar.
    - La matriz de datos estandarizada (X) y la llamada original (call).

Primero vamos a contextualizar los resultados. Se calculan los autovalores y autovectores de la matriz de covarianza de los datos (previamente estandarizados). Un autovalor representa la cantidad total de varianza que un componente principal captura de los datos originales. Siempre son positivos porque representan varianza, un concepto positivo. Los autovalores más grandes corresponden a los componentes principales que más información(variabilidad) contienen.  

En nuestro caso vemos que tenemos 7 componentes. Como comentamos anteriormente son todas las componentes posibles. Asociados a cada componente tenemos nuestros autovalores. El primer autovalor es 6.11 que explica un 87.3% de la variabilidad. Vamos que con esta primera componente ya explicaríamos la práctica totalidad de los datos. Después la variabilidad que explican los siguientes autovalores va reduciendose drásticamente autovalor a autovalor. De esta manera el segundo autovalor 0.39 explica un 5.57% de la variabilidad de los datos, el tercer autovalor 0.25 explica un 3.6%. El resto de autovalores son casi despreciables ya que explican apenas un 1%(si acaso) de la variabilidad de los datos. Como comentamos el valor del autovalor va directamente relacionado con la variabilidad que explica.

En el siguiente apartado vamos a determinar el número óptimo de componentes principales pero claramente podemos pronosticar que será 1. Lo veremos en detalle a continuación.


**Nota**: aunque nos refiramos a que el autovalor explique un tanto % de la variabilidad o varianza de los datos realmente es un abuso de notación o de forma de explicarnos. La que explica la variabilidad es la componente a la que está asociado el autovalor. El autovalor 1 está asociado a la componente 1 y así sucesivamente.

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

Tenemos que determinar el número óptimo de componentes principales: método de Kaiser, porcentaje de varianza y scree plot.

Método de Kaiser y scree plot.

```r
# Instalamos la librería "factoextra"
install.packages("factoextra")
```

```r
# Cargamos la librería "factoextra"
library(factoextra)

# Aplicamos el criterio del método de Kaiser
fviz_eig(PCA, choice = "eigenvalue", geom = "bar") +
  theme_grey() +
  geom_hline(yintercept = 1, colour = "red", linetype = "dashed", linewidth = 1)
```

 El gráfico muestra el scree plot de los autovalores del ACP, junto con la línea roja en λ=1 que representa el criterio de Kaiser pudiendo destacar los siguientes resultados:

+ El primer componente tiene un autovalor cercano a 6, muy por encima de 1, lo que indica que explica por sí solo una gran parte de la varianza total del conjunto de variables. En el apartado anterior vimos que explica el 87.3% de la varianza.

+ A partir del segundo componente todos los autovalores son claramente inferiores a 1 y la barra desciende de forma suave, sin grandes saltos, lo que sugiere que esas dimensiones capturan sobre todo ruido o variación específica, no factores latentes fuertes.

Siguiendo el criterio de Kaiser y la forma del scree plot, la interpretación razonable es que deberíamos conservar únicamente la primera componente ya que concentra la mayor parte de la información relevante de la base de datos.

Método del porcentaje de varianza explicada.

```r
# Tabla de autovalores y porcentajes de varianza (igual que PCA$eig)
eig_tab = get_eigenvalue(PCA)

# Extraemos la varianza acumulada
var_eig = data.frame(dim = 1:nrow(eig_tab), cum = eig_tab[, "cumulative.variance.percent"])

# Número óptimo de componentes según el criterio del porcentaje de varianza explicada (fijamos en 75%)
umbral = 75
r_opt = which(var_eig$cum >= umbral)[1]

# Aplicamos el criterio del método del porcentaje de varianza explicada
fviz_eig(PCA, choice = "variance", geom = "line") %+% var_eig +
  aes(x = dim, y = cum) +
  geom_point() +
  geom_line() +
  geom_hline(yintercept = umbral, linetype = "dashed") +
  geom_vline(xintercept = r_opt,  linetype = "dashed") +
  scale_x_continuous(breaks = var_eig$dim) +
  coord_cartesian(ylim = c(0, 100)) +
  labs(title = "Selección por porcentaje de varianza explicada",
           x = "Número de componentes (k)",
           y = "Varianza explicada acumulada (%)",
    subtitle = paste("Umbral = ", umbral, "%  →  r óptimo =", r_opt)) +
  theme_grey()
```

En este gráfico se representa la varianza explicada acumulada por los componentes principales, fijando un umbral del 75% (línea horizontal discontinua):

* La primera componente explica alrededor del 87% de la varianza total por encima claramente del nivel marcado como umbral.

* A partir de la segunda componente la curva sigue creciendo, pero con incrementos cada vez menores (ganancias marginales de información).

La línea vertical discontinua en k=1 indica que, según este criterio (conservar el mínimo número de componentes que explique al menos el 75% de la varianza), el número óptimo de componentes es 1. Con solo una componente ya capturamos casi el 90% de la variabilidad del sistema de variables, por lo que incluir más componentes no sería imprescindible si ese nivel de explicación se considera suficiente.

Como vemos los resultados concuerdan con la predicción que hicimos en el apartado 4 tras analizar los autovalores y la variabilidad explicada de cada componente.
